### Marketing AB Testing for Ad Campaign

Marketing companies want to run successful campaigns, but the market is complex and several options can work. So normally they tun A/B tests, that is a randomized experimentation process wherein two or more versions of a variable (web page, page element, banner, etc.) are shown to different segments of people at the same time to determine which version leaves the maximum impact and drive business metrics.

The companies are interested in answering two questions:

##### 1. Would the campaign be successful?
##### 2. If the campaign was successful, how much of that success could be attributed to the ads?

With the second question in mind, we normally do an A/B test. The majority of the people will be exposed to ads (the experimental group). And a small portion of people (the control group) would instead see a Public Service Announcement (PSA) (or nothing) in the exact size and place the ad would normally be.

The idea of the dataset is to analyze the groups, find if the ads were successful, how much the company can make from the ads, and if the difference between the groups is statistically significant.

Source: https://www.kaggle.com/code/simayakpinar/a-b-test-for-a-marketing-ad-campaign

```python
# Exploratory Data Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def perform_eda(csv_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file)
    
    # Display basic information about the DataFrame
    print("Basic Info:")
    print(df.info())
    
    # Display summary statistics of numerical columns
    print("\nSummary Statistics:")
    print(df.describe())
    
    # Display the first few rows of the DataFrame
    print("\nFirst Few Rows:")
    print(df.head())
    
    # Plot distribution of 'test group'
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x='test group')
    plt.title("Distribution of Test Groups")
    plt.xlabel("Test Group")
    plt.ylabel("Count")
    plt.show()
    
    # Plot conversion rate by test group
    plt.figure(figsize=(8, 6))
    sns.barplot(data=df, x='test group', y='converted', estimator=lambda x: sum(x) / len(x))
    plt.title("Conversion Rate by Test Group")
    plt.xlabel("Test Group")
    plt.ylabel("Conversion Rate")
    plt.show()
    
    # Plot distribution of 'total ads'
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x='total ads', bins=20, kde=True)
    plt.title("Distribution of Total Ads")
    plt.xlabel("Total Ads")
    plt.ylabel("Count")
    plt.show()
    
    # Plot distribution of 'most ads day'
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x='most ads day', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
    plt.title("Distribution of Most Ads Day")
    plt.xlabel("Most Ads Day")
    plt.ylabel("Count")
    plt.show()
    
    # Plot distribution of 'most ads hour'
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x='most ads hour', bins=24, kde=True)
    plt.title("Distribution of Most Ads Hour")
    plt.xlabel("Most Ads Hour")
    plt.ylabel("Count")
    plt.show()

# Example usage:
csv_file = "marketing_AB.csv"
perform_eda(csv_file)
```

Actual AB Testing

```python
# AB testing and answering the two questions

import pandas as pd

# Load the dataset
df = pd.read_csv("marketing_AB.csv")

# Calculate overall conversion rate
total_converted = df['converted'].sum()
total_population = len(df)
overall_conversion_rate = total_converted / total_population

# Calculate conversion rates for ad and PSA groups
ad_group = df[df['test group'] == 'ad']
psa_group = df[df['test group'] == 'psa']

conversion_rate_ad = ad_group['converted'].mean()
conversion_rate_psa = psa_group['converted'].mean()

# Calculate average number of ads seen for converted and non-converted individuals
avg_ads_seen_converted = df[df['converted'] == True]['total ads'].mean()
avg_ads_seen_not_converted = df[df['converted'] == False]['total ads'].mean()

# Output results
print("1. Would the campaign be successful?")
print("Overall Conversion Rate: {:.2%}".format(overall_conversion_rate))
if overall_conversion_rate > 0.5:
    print("Yes, the campaign would likely be successful.")
else:
    print("No, the campaign might not be successful.")

print("\n2. If the campaign was successful, how much of that success could be attributed to the ads?")
print("Conversion Rate for Ad Group: {:.2%}".format(conversion_rate_ad))
print("Conversion Rate for PSA Group: {:.2%}".format(conversion_rate_psa))
if conversion_rate_ad > conversion_rate_psa:
    print("The success could be attributed partially to the ads.")
else:
    print("The success could not be attributed to the ads.")

# Additional analysis: Relationship between ads seen and conversion rate
print("\nAdditional Analysis:")
print("Average number of ads seen by converted individuals: {:.2f}".format(avg_ads_seen_converted))
print("Average number of ads seen by non-converted individuals: {:.2f}".format(avg_ads_seen_not_converted))
```

#### Hypothesis Testing

```python
import pandas as pd
import scipy.stats as stats

# Load the dataset
df = pd.read_csv("marketing_AB.csv")

# Separate data into ad and PSA groups
ad_group = df[df['test group'] == 'ad']
psa_group = df[df['test group'] == 'psa']

# Calculate the number of conversions and total observations for each group
converted_ad = ad_group['converted'].sum()
total_ad = len(ad_group)

converted_psa = psa_group['converted'].sum()
total_psa = len(psa_group)

# Perform two-sample z-test for proportions
alpha = 0.05  # significance level

# Calculate pooled proportion
pooled_proportion = (converted_ad + converted_psa) / (total_ad + total_psa)

# Calculate standard error
SE = (pooled_proportion * (1 - pooled_proportion) * ((1 / total_ad) + (1 / total_psa))) ** 0.5

# Calculate test statistic
z_score = (converted_ad / total_ad - converted_psa / total_psa) / SE

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # two-tailed test

# Output results
print("Null Hypothesis Test:")
print("H0: Conversion rates of ad and psa groups are equal")
print("Ha: Conversion rates of ad and psa groups are not equal")

print("\nZ-score:", z_score)
print("P-value:", p_value)

if p_value < alpha:
    print("\nReject the null hypothesis. There is sufficient evidence to suggest that the conversion rates of ad and psa groups are different.")
else:
    print("\nFail to reject the null hypothesis. There is not enough evidence to suggest that the conversion rates of ad and psa groups are different.")
```
